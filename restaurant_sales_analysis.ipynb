```python
# ================================
# 1. IMPORT LIBRARIES
# ================================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# ================================
# 2. LOAD DATASET
# ================================
df1=pd.read_csv('/content/drive/MyDrive/AI_training/i1.csv')
df1.head()
```

```python
#2nd FILE
df2=pd.read_csv('/content/drive/MyDrive/AI_training/i2.csv')
```

```python
#merging two files
df = pd.concat([df1, df2], ignore_index=True)
display(df.head(10))
```

```python
# Displaying merged files
df.to_csv("merged.csv", index=False)
```

```python
# ================================
# 3. DATA OVERVIEW
# ================================
df.shape
df.columns
df.info()
df.describe()
df.isnull().sum()
df.duplicated().sum()
```

```python
# ================================
# 5. INVOICE-LEVEL ANALYSIS
# ================================
total_final_amount = df['Final Total'].sum()
num_unique_invoices = df['Invoice No.'].nunique()
avg_final_total_per_invoice = total_final_amount / num_unique_invoices

print(f"Total Final Amount: {total_final_amount}")
print(f"Number of Unique Invoices: {num_unique_invoices}")
print(f"Average Final Total per Invoice: {avg_final_total_per_invoice}")
```

```python
#invoice unique
invoice_count = df["Invoice No."].nunique()
```

```python
#find sum of all revenue
final_total_sum = df["Final Total"].sum()
```

```python
#average
average_per_invoice = final_total_sum / invoice_count

print("Final Total:", final_total_sum)
print("No of Invoices:", invoice_count)
print("Average per Invoice:", average_per_invoice)
```

```python
# revenue wise top 3 total
top_3_items = (
    df.groupby("Item Name")["Final Total"]
      .sum()
      .sort_values(ascending=False)[0:3]
      #.head(3)
)

print(top_3_items)
```

```python
# Filter data for 14-09-2023
df_date = df[df["Date"] == "2023-09-14"]

#find invoice spend on 14-09-2023
invoice_totals = (
    df_date.groupby("Invoice No.")["Final Total"]
    .sum()
)
highest_invoice = invoice_totals.idxmax()
highest_amount = invoice_totals.sum()   # or max()

print(highest_invoice)
highest_amount
```

```python
#invoice total bill
invoice_totals = (
    df.groupby("Invoice No.")["Final Total"]
    .sum()
)

avg = invoice_totals.mean()
print(avg)

grter_mean = invoice_totals[invoice_totals > avg].count()
grter_mean
```

```python

# ================================
# 6. BUSINESS QUESTIONS & ANALYSIS
# ================================

# ------------------------------------------------
# Q1. On each date, which product was the top seller (by quantity)?
# ------------------------------------------------


daily_product_qty = df.groupby(['Date', 'Item Name'])['Qty.'].sum().reset_index()


idx = daily_product_qty.groupby('Date')['Qty.'].idxmax()
top_sellers = daily_product_qty.loc[idx]


plt.figure(figsize=(12,6))
sns.barplot(data=top_sellers, x='Date', y='Qty.', hue='Item Name')
plt.title('Top Selling Product by Quantity (Each Day)')
plt.xticks(rotation=45)
plt.show()
```

```python
# ------------------------------------------------
# Q2. Do high-quantity items always generate high revenue?
# ------------------------------------------------


item_summary = df.groupby('Item Name').agg({
'Qty.': 'sum',
'Final Total': 'sum'
}).reset_index()


plt.figure(figsize=(8,6))
plt.scatter(item_summary['Qty.'], item_summary['Final Total'])
plt.xlabel('Total Quantity Sold')
plt.ylabel('Total Revenue')
plt.title('Quantity vs Revenue')
plt.show()



```

```python
# ------------------------------------------------
# Q3. Which product contributes the highest percentage of total revenue?
# ------------------------------------------------


total_revenue = df['Final Total'].sum()


item_summary['Revenue %'] = (item_summary['Final Total'] / total_revenue) * 100


item_summary = item_summary.sort_values('Revenue %', ascending=False)


plt.figure(figsize=(10,6))
sns.barplot(data=item_summary.head(5), x='Revenue %', y='Item Name')
plt.title('Top 5 Products by Revenue Contribution (%)')
plt.show()

```
